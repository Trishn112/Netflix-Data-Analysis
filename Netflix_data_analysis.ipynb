CLEANING THE DATA AND HANLDING ALL THE MISSING VALUES

```python
data_duplicate=data.copy() #NEVER DELETE DATA FROM ORGINAL FILES,
```

```python
data_duplicate.head()
```

```python
data_duplicate.isnull().info()
```

```python
data_duplicate.isnull().sum()
```

MAJOR NULL VALUE ARE PRESENT IN DIRECTOR SO WE HANDLE IT FIRST

```python
data_duplicate['director'].fillna("Unknow",inplace=True)
data_duplicate.head(10)
```

NOW WE WILL HANDLE THE NULL VALUES PRESENT IN CAST 

```python
data_duplicate['cast'].isnull().sum()
```

```python
data_duplicate['cast'].fillna("Unkown",inplace=True)
data_duplicate.head(10)
```

NOW SAME WITH COUNTRY

```python
data_duplicate['country'].isnull().sum()
```

```python
data_duplicate['country'].fillna("Unknow",inplace=True)
data_duplicate.head()
```

NEXT DATA TO HANDLE IS DATE COLUMNS

```python
data_duplicate['date_added'].isnull().sum()
```

```python
#IT HAS VEEY FEW VALUES SO WE CAN DROPS THE DATA
data_duplicate.dropna(subset=['date_added'],inplace=True)
```

```python
data_duplicate['date_added'].isnull().sum()
```

NOW HANDLE DATA FOR RATING

```python
data_duplicate['rating'].isnull().sum()
```

```python
data_duplicate['rating'].fillna("Not Rated",inplace=True)
```

```python
data_duplicate.isnull().sum()
```

```python
data_duplicate.info()
```

COLUMN date_added DTYPE IS OBJECT IT SHOULD BE IN THE FORMAT OF DATETIME

```python
data_duplicate['date_added']=(data_duplicate['date_added'].astype(str).str.strip())
```

```python
data_duplicate['date_added']=pd.to_datetime(data_duplicate['date_added'])
```

```python
data_duplicate['date_added'].info()
```

```python
data_duplicate['year_added'] = data_duplicate['date_added'].dt.year
data_duplicate['month_added'] = data_duplicate['date_added'].dt.month 
```

```python
data_duplicate['duration_num']=data_duplicate['duration'].str.extract('(\d+)').astype(str)
```

```python
data_duplicate.info()
```

How much content on Netflix is Movies vs TV Shows?

```python
data_duplicate['type'].value_counts()
```

```python
plt.Figure(figsize=(6,4))
sns.countplot(data=data_duplicate,x='type')
plt.title("Movies VS TV Show")
plt.show()
```

NETFLIX HAS MORE MOVIES THAN TV SHOW 

How has Netflix content grown over time?

```python
data_duplicate['year_added'].value_counts().sort_index()
```

```python
plt.Figure(figsize=(10,5))
data_duplicate['year_added'].value_counts().sort_index().plot(kind='line')
plt.xlabel("year")
plt.ylabel("no. of tiles")
plt.title("NETFLIX GROWTH OVER YEAR")
```

Which countries produce the most Netflix content?

```python
countries=data_duplicate['country'].str.split(', ').explode()
top10=countries.value_counts().head(10)
top10
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x=top10.values,y=top10.index)
plt.ylabel("Countries")
plt.xlabel("Contents")
plt.show()
```

Major rated show on Netflix ?

```python
data_duplicate['rating'].value_counts()
```

```python
plt.Figure(figsize=(10,5))
data_duplicate['rating'].value_counts().plot(kind='bar')
plt.ylabel("Shows")
plt.show()
```

Typical length of Netflix movies?

```python
movies=data_duplicate[data_duplicate['type']=='Movie']
plt.figure(figsize=(90,45))
sns.histplot(movies['duration_num'],bins=30)
plt.xlabel('duration (min)',fontsize=24)
plt.ylabel('count',fontsize=24)
plt.show()
```

ALL RESULT IN PONITS AFTER THE WHOLE ANALYSIS :-

->The following significant findings were attained following data cleaning, exploratory data analysis, and visualization on  the Netflix dataset:

->Movies make up a far greater percentage of Netflix's content library than TV series do.

->After 2015, the platform's content addition rate skyrocketed, indicating aggressive expansion and worldwide growth.

->Netflix's emphasis on both Western and emerging markets is demonstrated by the fact that the United States contributes the most content, followed by India and the United Kingdom.

->The most popular genres are dramas and foreign films, demonstrating Netflix's focus on content that appeals to a wide audience and is emotionally compelling.

->It appears that Netflix primarily caters to adult viewers because the majority of its content is categorized under mature audience ratings like TV-MA and TV-14.
